```python
def greet(): # function definition # define a function named greet # 
    print("Hello Vishwas") # function body # code to be executed when the function is called

greet() # function calling # when we call the function, the code inside the function will be executed
greet() # function calling
greet() # function calling
```

```python

# Parameter vs Argument

# Parameter: A parameter is a variable in the function definition.
# Argument: An argument is the value passed to the function when it is called.

def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

greet("Alice")  # "Alice" is an argument
greet("Bob")    # "Bob" is an argument
```

```python
# multiple Parameters:- multiple parameters are separated by commas in the function definition.


def add(a,b): # function definition with two parameters a and b
    sum = a+b # code to add a and b and store the result in sum
    return sum # return the value of sum to the caller

add(5,6)   # function calling 
```

```python
# Default Argument :- A default argument is a parameter that has a default value.
#  If the caller does not provide a value for that parameter, the default value will be used.

def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Vishwas")
```

```python
# Keyword Argument :- A keyword argument is an argument that is passed to a function by explicitly
#                      specifying the parameter name.

def student(name, age):
    print(name, age)

student(age=20, name="Vishwas")
```

```python
# positional argument:- A positional argument is an argument that is passed to a function without 
# explicitly specifying the parameter name. The order of the arguments matters in this case.

def student(name, age):
    print(name, age)

student("Vishwas",20)
```

```python
# Arbitrary Arguments:- Sometimes, we may not know in advance how many arguments a function will receive.
#  In such cases, we can use arbitrary arguments.

def numbers(*args):
    print(args)

numbers(10,20,30,40)
```

```python
def total(*args):
    print(sum(args))

total(10,20,30)
```

```python
# Arbitary Keyword Arguments:- Sometimes, we may not know in advance how many keyword arguments 
#    a function will receive.   In such cases, we can use arbitrary keyword arguments.

def info(**kwargs):
    print(kwargs)

info(name="Vishwas", age=20)
```

```python
# Local Variable :- A local variable is a variable that is defined inside a function
#  and can only be accessed within that function.

def greet():
    name = "Vishwas" # local variable
    print("Hello", name)

greet()
```

```python
# Global Variable :- A global variable is a variable that is defined outside of any 
# function and can be accessed from anywhere in the program.


x = 100 # global variable

def test(): # function definition
    print(x) # access the global variable x inside the function

test() # function calling
```

```python
# Global Keyword :- The global keyword is used to declare a variable as global inside a function. 
# This allows us to modify the value of a global variable from within a function.

x = 100 # global variable

def test():
    global x # declare x as global
    x = 200 # modify the global variable x
    print(x)

test()
print(x)
```

```python
# Nested Function :- A nested function is a function that is defined inside another function.
def outer():
    print("This is the outer function.")
    
    def inner():
        print("This is the inner function.")
    
    inner() # call the inner function inside the outer function 

outer() # call the outer function
```

```python
# Recursion :- A recursive function is a function that calls itself in order to solve a problem.
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1) 
print(factorial(5))
```

```python
# lambda function:- A lambda function is a small anonymous function that can take any
#  number of arguments, but can only have one expression. It is defined using the lambda keyword. 
# syntax --->
# lambda arguments: expression

add = lambda a, b: a + b
print(add(5, 6))
   
```

```python
# map() function:- The map() function is a built-in function in Python that applies a given function
#  to each item of an iterable (like a list) and returns a map object (which is an iterator).

nums = [1,2,3,4]

result = list(map(lambda x:x*x, nums))

print(result)
```

```python
# filter() function:- The filter() function is a built-in function in Python that constructs
#  an iterator from elements of an iterable for which a function returns true.

nums = [1,2,3,4,5,6]

result = list(filter(lambda x:x%2==0, nums))

print(result)
```

```python
# reduce() function:- The reduce() function is a built-in function in Python that applies a
#  binary function cumulatively to the items of an iterable, from left to right, to reduce the 
# iterable to a single value. It is defined in the functools module,
#  so we need to import it before using.

from functools import reduce
nums = [1,2,3,4]
result = reduce(lambda x,y:x+y, nums)
print(result)
```

```python

# Docstring:- A docstring is a string literal that is used to document a function, class, or module.

def add(a,b):
    """
    This function adds two numbers
    """
    return a+b

print(add.__doc__)
```

```python
# generator function:- A generator function is a special type of function that returns an iterator.

def count():

    yield 1
    yield 2
    yield 3

x = count()

print(next(x))
print(next(x))
print(next(x))
```

```python
# Decorator:- A decorator is a function that takes another function as an argument and extends 
# its behavior without modifying it.

def decorator(func):

    def wrapper():
        print("Before")

        func()

        print("After")

    return wrapper


@decorator
def hello():
    print("Hello")

hello()
```

```python
""" Q6. (Arbitrary Arguments: *args) Write a function named total_marks() that accepts any number 
of subject marks using *args. Inside the function, use a loop or the sum() function to calculate the 
total of all the marks passed, and return the final score. Try calling it with 3 subjects, and then
 with 5 subjects."""

def total_marks(*args):
    total = 0

    for mark in args:
        total += mark

    return total

print(total_marks(80, 90, 85))
print(total_marks(80, 90, 85, 75, 95))

    
```

```python
""" Q8. (Keyword Arbitrary Arguments: kwargs) Create a function named character_profile() that takes 
kwargs. This allows you to pass any number of keyword arguments (like name="Gojo", 
domain="Unlimited Void", power_level=999). Inside the function, use a for loop with .items()
 to print out all the key-value pairs beautifully.
"""
def character_profile(**kwargs):

    for key , values in kwargs.items():
        print(f"{key} : {values}")

character_profile(
    name="Gojo",
    domain="Unlimited Void",
    power_level=999
)



```

```python
""" Q9. (Combining Argument Types) Write a complex function named attack_sequence()
 that takes parameters in this exact order:
attacker (Regular positional argument)
target (Regular positional argument)
*moves (Arbitrary positional arguments for attack names)
damage_multiplier (Keyword argument with a default value of 1.0)"""

def attack_sequence(attacker, target, *moves, damage_multiplier=1.0):
    print(f"Attacker: {attacker}")
    print(f"Target: {target}")

    print("\nMoves Used:")
    for move in moves:
        print("-", move)

    print(f"\nDamage Multiplier: {damage_multiplier}")


attack_sequence(
    "Gojo",
    "Sukuna",
    "Blue",
    "Red",
    "Hollow Purple",
    damage_multiplier=2.5
)
```

```python
""" Q10. (Logic & Docstrings) Write a function called verify_password() that takes a password string.
Add a Docstring (using triple quotes "") at the top of the function explaining what it does.
The function should return True if the password has at least 8 characters AND contains the symbol "@".
Otherwise, it should return False.
Test it with passwords like "kali" and "admin@12345".
"""

def verify_password(password):
    """
    Checks whether a password is valid.
    A valid password must have at least 8 characters
    and contain the '@' symbol.
    """

    if len(password) >= 8 and "@" in password:
        return True
    else:
        return False


print(verify_password("kali"))
print(verify_password("admin@12345"))
```
