```python
# >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>...STRING...<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<

""" A string is simply a piece of text treated as a single unit, 
    created by surrounding characters with quotes (like "Hello" or 'Python').
 """
```

```python
# Example of string:- 
name = "Vishwas"
city = 'Jaipur'
```

```python
a = "Hello"
print(a)
```

```python
# String indexing :- string indexing allows you to access a specific character by its position,
#                    starting at 0 for the first character.

text = "python"

print(text[0]) 
print(text[1]) 
print(text[4]) 
print(text[3]) 
print(text[2])
print(text[-1])
print(text[-3])
```

```python
# String slicing :- String slicing is the process of extracting a specific portion (substring) from a string

#syntax :- 
# string[start : end : step]
```

```python
# Slicing example :-->
text = "PYTHON"

print(text[0:3])  # basic slicing
print(text[:4])   # Without start
print(text[2:])   # without end
print(text[:])    # full string
print(text[0:6:2]) # step slicing
print(text[::-1])  # reverse string
```

```python
# Concatenation 

a = "Hello"
b = "World"

print(a + b) # simple
print(a + " " + b) # with space
```

```python
# Repetition
print("Hi " * 3)
```

```python
# Membership Operator
text = "Python"

print("Py" in text)
print("Java" not in text)
```

```python
# len() Function :--> to find the length of string

text = "Python"

print(len(text))
```

```python
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>...String Methods...<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
```

```python
#lower() :- The lower() method turns all the letters in a string into lowercase (small letters).

text = "PYTHON"
print(text.lower())
```

```python
# upper():- The `upper()` method turns all the letters in a string into **uppercase** (capital letters).

print("python".upper())
```

```python
# title() :- The title() method capitalizes the first letter of every word in a string and makes the rest of the letters lowercase.

text = "hello world"
print(text.title()) 
```

```python
# capitalize() :- """The capitalize() method turns only the very first letter of the string 
#                     into an uppercase letter and makes all other letters lowercase."""

text = "hello WORLD"
print(text.capitalize()) 
```

```python
# strip() :- The strip() method removes all leading and trailing whitespace (spaces, tabs, and newlines) from a string.

text = "  hello  "
print(text.strip())
```

```python
# replace() :- The replace() method searches for a specific part of a string and replaces it with a new piece of text.

text = "I like Java"
print(text.replace("Java", "Python"))
```

```python
# find() :- The find() method searches for a specific substring and returns the index 
#           of its first occurrence; if the substring is not found, it returns -1.

text = "Python"
print(text.find("t"))
```

```python
# count():-The count() method returns the total number of times a specific substring or character appears in a string.

text = "banana"
print(text.count("a"))
```

```python
# startswith():- The startswith() method checks if a string begins with a specific
#                 character or phrase and returns True if it does, otherwise False.

text = "Python is great"
print(text.startswith("Python")) 
```

```python
# endswith() :- The endswith() method checks if a string ends with a specific 
#               character or phrase and returns True if it does, otherwise False

text = "hello.py"
print(text.endswith(".py")) 
```

```python
# split() :- The split() method breaks a string into a list of smaller strings based on a 
#            specific separator (like a space, comma, etc.).
#            If no separator is specified, it splits by any whitespace.

text = "Apple,Banana,Cherry"
print(text.split(","))
```

```python
# join() :-   The join() method takes all items in an iterable (like a list) 
#             and joins them into one string, using a specified character as a separator.

words = ["Python", "is", "fun"]
result = " ".join(words)
print(result)
```

```python
# Escape Characters :- Escape characters use a backslash (\) to insert special formatting or
#                       characters that are otherwise hard to write inside a string.

print("Hello\nWorld") # new line
print("Hello\tWorld") # tab space
print("He said \"Hello\"") # print quotes
```

```python
# f-String Formatting :- f-strings (formatted string literals) allow you to embed variables and expressions directly 
#                         inside a string by prefixing it with an f and placing the content inside curly braces {}.

name = "Vishwas"
age = 20
print(f"My name is {name} and age is {age}")
```
