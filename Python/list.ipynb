```python
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>...LIST...<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
"""
A List in Python is an ordered, mutable (changeable) collection of items that can
hold different data types, created by placing elements inside square brackets [].
"""
# Key Characteristics
"""
Ordered: Items have a defined order, and that order will not change unless explicitly modified.

Mutable: You can add, remove, or change items after the list is created.

Allows Duplicates: You can have the same item multiple times in a list.

Heterogeneous: A single list can contain integers, strings, objects, and even other lists.  """
```

```python
# example :--->
data = [10, "Vishwas", 5.5, True]
print(data)
```

```python
# LIST INDEXING :-  List indexing allows you to access individual items in a
#                   list using their position, starting at 0 for the first element.

fruits = ["Apple", "Banana", "Mango"]
print(fruits[0])  #Positive Indexing
print(fruits[-1]) #Negative Indexing
```

```python
# List Slicing :- List slicing is the process of extracting a specific portion (sub-list) from a list

# Syntax :-->
# list[start:end].
```

```python
# Examples :- 
nums = [10,20,30,40,50]
print(nums[1:4])   # Simple
print(nums[:3])    # From start
print(nums[2:])    # last tak
print(nums[:])     # Full copy
print(nums[::2])   # Step slicing
print(nums[::-1])  # Reverse
```

```python
# Updating List:- 

fruits = ["Apple", "Banana", "Mango"]
fruits[1] = "Orange"
print(fruits)
```

```python
# Adding elements :- 

#1. append():- adds a single element to the very end of an existing list.

nums = [10,20]
nums.append(30)
print(nums)

#2. insert():- adds an element at a specific position in a list, pushing the existing elements to the right.

nums.insert(1,15)
print(nums)

#3. extend() :- dds all the elements from an iterable (like another list, tuple, or set) to the end of the current list.

a = [1,2]
b = [3,4]
a.extend(b)
print(a)

"""Key Difference
Unlike  append(), which adds the entire object as a single element,
        extend() "unpacks" the iterable and adds each item individually.
"""
```

```python
# Removing Elements :-

#1. remove() :- The remove() method searches for the first occurrence of a specified value and deletes it from the list.

nums = [10,20,30]
nums.remove(20)
print(nums)

#2. # pop() :- The pop() method removes and returns the element at a specific index. 
#          If you don't provide an index, it removes and returns the last element of the list.

nums = [10,20,30]
nums.pop(1)
print(nums)

#3. del :- The del keyword is a Python statement used to delete objects. In the context of lists, it removes an item at
#            a specific index or deletes an entire slice. Unlike pop(), it does not return the deleted item.

nums = [10,20,30]
del nums[1]
print(nums)

#4. clear() :- The clear() method removes all elements from a list, leaving it as an 
#              empty list [] but keeping the list object itself in memory. 

nums = [10,20]
nums.clear()
print(nums)
```

```python
#1. len() :- The len() function returns the total number of items currently 
#          inside a list (or any other collection like strings, tuples, or dictionaries).

nums = [10,20,30]
print(len(nums))

#2. max() :- The max() function returns the largest item in an
#            iterable (like a list) or the largest of two or more arguments.

print(max(nums))

#3. min() :- The min() function returns the smallest item in an iterable (like a list) or the smallest of two or more arguments.

print(min(nums))

#4. sum() :- The sum() function calculates the total sum of all numerical items in an iterable (like a list, tuple, or set).

print(sum(nums))
```

```python
# Sorting

# Sort() :- The sort() method sorts the items of a list in place, meaning it modifies the original list
#            directly rather than creating a new one. By default, it sorts in ascending order.

nums = [50,10,30]
nums.sort()
print(nums)

# Descending :- To sort a list in descending order (from largest to smallest), 
#               you can use the reverse=True argument with the sort() method or the sorted() function.

nums.sort(reverse=True)
print(nums)
```

```python
# Reverse :- To reverse a list in Python, you can use either the reverse()
#             method (to change the original list) or slicing (to create a new, reversed copy).

nums = [10,20,30]
nums.reverse()
print(nums) 
```

```python
#Count() :- The count() method returns the total number of times a specific value (element) appears in a list.

nums = [10,20,10,30]
print(nums.count(10))
```

```python
# index() method :- The index() method is used to find the first position (index) of a specific value within a list.

nums = [10,20,30]
print(nums.index(20))
```

```python
# Copy() :- The copy() method returns a shallow copy of the list. This means it creates a new list object 
#            that contains the same elements as the original, but the two lists remain independent.

a = [10,20,30]
b = a.copy()
print(b)
```
