```python
#>>>>>>>>>>>>...Mutable Aur Immutable Object...<<<<<<<<<<<<<<<
```

```python
# Immutable Object :- that objects cannot be change

# example :- int, float, str, tuple, bool
```

```python
 # Immutable Objects Example

a = 10          # int
b = 3.14        # float
c = "Python"    # str
d = (1, 2, 3)   # tuple
e = True        # bool

print(type(a), a)
print(type(b), b)
print(type(c), c)
print(type(d), d)
print(type(e), e)
```

```python
# Mutable Object :- that object can be change 

# example :- list, dict, set
```

```python
# mutable Objects Example
numbers = [1, 2, 3]

print("Before:", numbers)
print("ID:", id(numbers))

numbers.append(4)

print("After :", numbers)
print("ID:", id(numbers))
```
