```python
# 1. FOR LOOP :- A loop that steps through a sequence item by item until it reaches the end.

""" 
SYNTAX :--->
for item_variable in sequence:
    # Code to run for each item
"""
# Example :-

for number in range(3):
    print("Hello!") 
```

```python
# Real-World Example: Sending Bulk Data
user_emails = ["admin@bkriet.com", "student@bkriet.com"]
for email in user_emails:
    print("Sending welcome email to:", email)
```

```python
# 2. WHILE LOOP :- A loop that keeps repeating as long as a certain condition remains True.
""" 
SYNTAX :--->
while condition_is_true:
    # Code to execute
"""
# Example :-
count = 1

while count <= 3:
    print(count)
    count = count + 1
```

```python
password = ""

while password != "secret123":
    # In a real program, we would use input() to ask the user. 
    # Here we simulate the user eventually typing it right.
    password = "secret123" 
    print("Access Granted.")
```

```python
# Break Statement :- An emergency exit for a loop.

# Example :-

for num in range(1, 6): # Numbers 1, 2, 3, 4, 5
    if num == 3:
        print("Found 3! Stopping early.")
        break
    print("Checking number:", num)
```

```python
# Continue Statement :- Skips the current turn of the loop and jumps to the next one.

for num in range(1,11):
    if num == 3:
        print("skip")
        continue
    print(num)
```

```python
# Nested loop :- A loop placed inside another loop.
""" 
SYNTAX :--->
for item in outer_collection:
    for sub_item in inner_collection:
        # Code to execute
"""
# Example :- 

colors = ['Red', 'Blue']
items = ['Car', 'Bike']
for color in colors:
    for item in items:
        print(color, item)
```

```python
""" Q1. (Basic for loop) Use a for loop and the range() 
function to print numbers from 1 to 10 on separate lines.
"""
for n in range(1,11):
    print(n)
```

```python
""" 
Q2. (Basic while loop) Create a variable count = 1. Use a while loop to print 
the first 5 multiples of 3 (3, 6, 9, 12, 15).
"""
count = 1
while  count < 16:
    count += 1
    if count%3 == 0:
        print(count)
```

```python
""" Q3. (Looping through a List) You have a list of marks: marks = [85, 90, 75, 95, 80]. 
Use a for loop to calculate and print the total sum of all the marks in the list.
"""
marks = [85, 90, 75, 95, 80]
sum = 0
for mark in marks:
    sum += mark
print(sum)
```

```python
""" Q4. 
(Using break) You have a list of names:
 anime_list = ["Naruto", "Jujutsu Kaisen", "Solo Leveling", "Demon Slayer"]. 
Loop through the list and print each name. However, if the loop encounters
 "Solo Leveling", print "Target found!" and stop the loop immediately using the break statement.
"""
anime_list = ["Naruto", "Jujutsu Kaisen", "Solo Leveling", "Demon Slayer"]

for anime in anime_list:
    print(anime)
    if (anime == "Solo Leveling"):
        print("Target found!")
        break
```

```python
""" 
Q5. (Using continue) Write a for loop that iterates from number 1 to 10. If the number is 5 or 7,
 use the continue statement to skip printing it. Your final output should show all numbers 
 from 1 to 10 except 5 and 7.
"""
for i in range(1,11):
    if i == 5:
        continue
    elif i == 7:
        continue
    print(i)
```

```python
""" Q6. (While loop with condition) Create a variable countdown = 5. 
Use a while loop to print the numbers decreasing from 5 down to 1. 
After the loop finishes, print "Launch Successful!".
"""

countdown = 5

while 1 <= countdown <= 5:
    print(countdown)
    countdown -= 1
```

```python
""" Q7. (Looping through a String) Create a string variable word = "Python". 
Use a for loop to print each character of the string, but use a continue
statement to skip printing the letter "h". 
"""

word = "Python" 
for qr in word:
    if(qr == "h"):
        continue
    print(qr)
```

```python
""" Q8. (Nested Loops) Write a nested loop (a loop inside another loop) to print a pattern of
 stars (*). The final output in your notebook cell should look exactly like this:
*
**
***
****
*****
"""

for i in range(1,6):
    for j in range(1):
        print(i * "*")
```

```python
""" Q9. (Simulating a Login Limit) Create a while True: loop (an infinite loop). Inside the loop,
increment a variable named attempts starting from 1. Every time the loop runs,
print "Attempt number: X" (replace X with the attempt number). If attempts reaches 3,
 print "Account Locked!" and use the break statement to stop the loop safely.
"""

while True:
    attempts = 1
    while attempts <=3:
        print(f"Attempt number :{attempts}")
        attempts += 1
    print("Account Locked!")
    break
```
