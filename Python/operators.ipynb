```python
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>...OPERATORS...<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
"""
Python operators are special symbols used to perform operations on variables and values
"""
# example :- 
a = 10
b = 5
print(a + b) # + is a operator
```

```python
# 1. Arithmetic Operators :- Used to perform mathematical calculations.


a = 15
b = 4

print(f"Values: a = {a}, b = {b}\n")

print(f"Addition (+):       {a + b}")
print(f"Subtraction (-):    {a - b}")
print(f"Multiplication (*): {a * b}")
print(f"Division (/):       {a / b}")        
print(f"Floor Division (//): {a // b}")      
print(f"Modulus (%):        {a % b}")        
print(f"Exponent (**):      {a ** b}")       
```

```python
# Operator Precedence :- Operator Precedence is the set of rules that determines the order 
#                         in which Python evaluates different operators in an expression.


# 1. Without Parentheses:- 
# Calculation: 5 + (2 * 10) = 5 + 20 = 25
result1 = 5 + 2 * 10

# 2. With Parentheses 
# Calculation: (5 + 2) * 10 = 7 * 10 = 70
result2 = (5 + 2) * 10

print(f"Result 1 (Without brackets): {result1}")
print(f"Result 2 (With Brackets): {result2}")
```

```python
# 2. Comparision Operators :- Comparison operators are symbols used to compare two values,
#                               returning True if the condition is met and False otherwise.


a = 10
b = 5

print(f"Equal (==):        {a == a}")    
print(f"Not Equal (!=):    {a != b}")    
print(f"Greater Than (>):  {a > b}")     
print(f"Less Than (<):     {b < a}")     
print(f"Greater Equal (>=):{a >= 10}")   
print(f"Less Equal (<=):   {b <= 10}")   

# Chained Comparison
age = 20
print(f"Chained (18 <= age <= 60): {18 <= age <= 60}") 
```

```python
# 3. Assignment Operators :- Assignment operators are used to assign values to variables,
#                              often performing an operation simultaneously.

x = 10

x += 5   # x = 10 + 5
print(f"x += 5: {x}")   

x -= 3   # x = 15 - 3
print(f"x -= 3: {x}")   

x *= 2   # x = 12 * 2
print(f"x *= 2: {x}")   

x /= 4   # x = 24 / 4
print(f"x /= 4: {x}")   

x %= 4   # x = 6.0 % 4
print(f"x %= 4: {x}")   

x **= 3  # x = 2.0 ** 3
print(f"x **= 3: {x}")  
```

```python
# 4. Logical Operators :- Logical operators are used to combine conditional statements
#                         and evaluate the relationship between them, returning True or False.

x = 10

# AND: Both must be true
print(x > 5 and x < 15) 

# OR: At least one must be true
print(x > 5 or x < 5)    

# NOT: Reverses the condition
print(not(x > 5))       

#Short Circuiting :- don't check unnecessary condition 

False and print("Hello1")
True and print("Hello11")

True or print("Hello2")
False or print("Hello22")
```

```python
# 5. Identity Operators :- to compare memory address

a = [1, 2]
b = a
c = [1, 2]

# 'is' operator: Memory address compare karta hai
print(f"a is b: {a is b}")      
print(f"a is c: {a is c}")     

# 2. 'is not' operator
print(f"a is not c: {a is not c}") # True

# 3. 'is' vs '=='
# '==' sirf value compare karta hai
# 'is' object (memory address) compare karta hai
print(f"a == c: {a == c}")     
print(f"a is c: {a is c}")      
```

```python
# 6. Membership Operators :- Membership operators are used to test if a value or variable is present 
#                             within a sequence (such as strings, lists, or tuples).

fruits = ["Apple","Mango","Banana"]
print("Mango" in fruits)


print("Orange" not in fruits)
```

```python
# 7. Bitwise Operators :- to perform operations on individual bits of integer numbers by comparing their binary representations.


a = 5  # Binary: 101
b = 3  # Binary: 011

print(f"AND (5 & 3): {a & b}")    # 001 = 1
print(f"OR  (5 | 3): {a | b}")    # 111 = 7
print(f"XOR (5 ^ 3): {a ^ b}")    # 110 = 6
print(f"NOT (~5)   : {~a}")       # -(5+1) = -6

# Shift Operators
print(f"Left Shift (5 << 1): {5 << 1}")   # 101 -> 1010 = 10
print(f"Right Shift (8 >> 1): {8 >> 1}")  # 1000 -> 100 = 4
```

```python
# 8. Ternary Operator :- The ternary operator is a shorthand way to write an if-else statement in a single line.

# SYNTAX :---->      value_if_true if condition else value_if_false


age = 20
result = "Adult" if age >= 18 else "Minor"
print(result)
```

```python
# 9. Operator Precedence :- It is the rule set that determines the order in which different operators are evaluated in an expression.
""" 
1. ( )
2. ** 
3. ~, +, - (unary operators)
4. *, /, //, %
5. +, -
6. <<, >>
7. &
8. ^
9. '
10. ==, !=, >, <, >=, <=, is, is not, in, not in
11. not
12. and 
13. or
"""
result = 10 + 2 ** 3 * 4 / 2 - 5
print(f"Result without parentheses: {result}")
 
result_with_parens = (10 + 2) ** 3 * 4 / 2 - 5
print(f"Result with parentheses: {result_with_parens}") 
```

```python
# Special Concepts 

# String Concatenation
print("Hello" + " World")

# String Repetition
print("Hi " * 3)


# List Concatenation
a = [1, 2]
b = [3, 4]
print(a + b)


# List Repetition
print([1, 2] * 3)
```
