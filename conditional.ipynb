```python
# 1. if Statement :- if statement is used to execute a block of code when a condition is True.

""" 
SYNTAX :---> 

if condition:
    code
"""
# example :- 

age = 18
if age >= 18:
    print("You can vote")     # indentation is important 
```

```python
# 2. if-else Statement :- It executes one block of code if the condition is True, otherwise it executes another block.
""" 
SYNTAX :---> 
if condition:
    # Code if condition is True
else:
    # Code if condition is False
"""
# example :-

age = 18
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
# 3. if-elif-else Statement :- is used to check multiple conditions and execute the matching block of code.
""" 
SYNTAX :---> 
if condition1:
    code
elif condition2:
    code
elif condition3:
    code
else:
    code
"""
# example :- 
marks = 85

if marks >= 90:
    print("A Grade")
elif marks >= 75:
    print("B Grade")
elif marks >= 50:
    print("C Grade")
else:
    print("Fail")
```

```python
# 4. Nested if Statement :- A Nested if is an if statement inside another if statement.
""" 
SYNTAX :---> 
if condition1:
    # Code block

    if condition2:
        # Code block
"""
# example :- 
age = 20

if age >= 18:
    print("Adult")

    if age >= 21:
        print("Can apply")
```

```python
# 5. Short Hand if :- 

age = 20
if age >= 18: print("Adult")
```

```python
# 6. Short Hand if-else :- 
""" 
SYNTAX :--->
value_if_true if condition else value_if_false
"""
# example :-
 
age = 20
result = "Adult" if age >= 18 else "Minor"
print(result)
```

```python
""" 
Q1. (Basic if-else) Create a variable number = -5. Write an if-else statement to check if the number is
positive (greater than or equal to 0) or negative. Print an appropriate message for each case.
"""
number = -5
if number >= 0:
    print("Positive number :", number)
else:
    print("Negative number :",number)    
```

```python
""" Q3. (Even or Odd) Take an integer variable num = 14. Use the modulus operator % 
and an if-else statement to check if the number is Even or Odd. Print "Even" or "Odd" accordingly.
"""
num = 14
if num%2 == 0:
    print("Even")
else:
    print("Odd")
        
```

```python
""" Q4. (Grading System) Create a variable marks = 82. Write an if-elif-else 
ladder to assign grades based on these rules:

Marks >= 90: Grade A

Marks >= 80 and < 90: Grade B

Marks >= 70 and < 80: Grade C

Marks < 70: Grade F
Print the final grade.
"""
Marks = 82

if Marks >= 90:
    print("Grade A")
elif (Marks >= 80) and (Marks < 90):
    print("Grade B")
elif (Marks >= 70) and (Marks < 80):
    print("Grade C")
else:
    print("Grade F")
```

```python
""" Q5. (Traffic Light Simulator) Create a string variable light = "Yellow". Write a conditional structure that checks the value of light:

If it is "Red", print "Stop!"

If it is "Yellow", print "Slow Down / Wait!"

If it is "Green", print "Go!"

If it is anything else, print "Invalid Light Color"
"""
light = "Yellow"

if light == "Red":
    print("Stop!")
elif light == "Yellow":
    print("Slow Down / Wait!")
elif light == "Green":
    print("Go!")
else:
    print("Invalid Light Color")
```

```python
""" Q6. (String Length Validation) Imagine a registration form where usernames must be at least 5
 characters long. Create a variable username = "ai". Write a code that checks the length of the 
 string using len(). If it's less than 5, print "Username too short".
 Otherwise, print "Username accepted".
"""
username = "ai"
length = len(username)
if length < 5:
    print("Username too short")
```

```python
""" Q7. (Largest of Two) Create two variables x = 45 and y = 65. Write an if-elif-else block to 
find out which number is larger, or if they are equal. Print a message like "x is greater", 
"y is greater", or "Both are equal".
"""
x = 45
y = 65

if x > y:
    print(f"{x} is greater")
elif y > x:
    print(f"{y} is greater")
else:
    print("Both are equal")
```

```python
"""
Q8. (Largest of Three) Create three variables: a = 12, b = 27, and c = 19. Use logical operators
 (and, or) inside your conditional statements to find and print the largest number among the three 
 without using the built-in max() function.
"""
a = 12 
b = 27
c = 19

if (a > b) and (a > c):
    print(f"{a} is greater among the three")
elif (b > a) and (b > c):
    print(f"{b} is greater among the three")
elif (c > a) and (c > b):
    print(f"{c} is greater among the three")
else:
    print("all are equal")
```

```python
""" 
Q9. (Leap Year Checker) Create a variable year = 2026. Write a program to check if the 
year is a leap year or not. (Hint: A year is a leap year if it is divisible by 4, but centurial
years like 1900, 2100 must also be divisible by 400 to be leap years).
"""
year = 2026 
if year%400 == 0:
    print("Leap Year")
elif year%100 == 0:
    print("not leap year")
elif year%4 == 0:
    print("Leap year")
else:
    print("not leap year")
```

```python
""" Q10. (Nested Login System Simulation) Create two variables to hold credentials:
saved_username = "admin"
saved_password = "password123"

Now, simulate a user trying to log in with these inputs:

entered_username = "admin"
entered_password = "wrong_password"

Write a nested if-else structure (an if inside an if) that:

First checks if the entered_username matches saved_username. If it doesn't match, print "User Not Found".

If the username matches, then check if the entered_password matches saved_password. If it matches, 
print "Access Granted". If it doesn't, print "Incorrect Password".
"""

saved_username = "admin"
saved_password = "password123"

entered_username = "admin"
entered_password = "wrong_password"

if entered_username == saved_username:
    if entered_password == saved_password:
        print("Access granted")
    else:
        print("Incorrect Password")
else:
    print("User Not Found")        

```
