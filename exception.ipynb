What is an Exception?
xception is an error that occurs during program execution.

```python
# using exception handling :-

a = 10
b = 0
try:
    print(a/b)

except:
    print("Error occurred")

print("Program End")
```

```python
# Try block :--->
""" 
syntax :-
try:
    risky code
"""
# except block :-->
""" 
syntax :-->
try:
    risky code

except:
    handling code
"""
# else block :-->
""" 
syntax :--->
try:
    code

except:
    code

else:
    code
"""

# finally block :-->
""" 
Syntax :-
try:
    code

except:
    code

finally:
    code"""
```

```python
try:
    print(10/0)

except:
    print("Cannot divide by zero")
```

```python
try:
    print(10/0)

except ZeroDivisionError:
    print("Division by zero not allowed")
```

```python
try:
    num = int(input())
    print(10/num)

except ValueError:
    print("Invalid Input")

except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# using exception object :-->
try:
    print(10/0)

except ZeroDivisionError as e:
    print(e)
```

```python
# Exception class :- all exception present in a class

try:
    print(10/0)

except Exception as e:
    print("Error:", e)
```

```python
try:
    print(10/2)

except:
    print("Error")

else:
    print("Success")
```

```python
try:
    print(10/0)

except:
    print("Error")

finally:
    print("Always Runs")
```

```python
# Nested Exception Handling
try:

    try:
        print(10/0)

    except ZeroDivisionError:
        print("Inner Exception")

except:
    print("Outer Exception")
```

```python
# Rising exception :--> using raise keyword
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```

```python
# Custom Exception 
class InvalidAgeError(Exception):
    pass
```

```python
age = -5

if age < 0:
    raise InvalidAgeError("Invalid Age")
```

```python
class LowBalanceError(Exception):
    pass

balance = 500

if balance < 1000:
    raise LowBalanceError("Minimum balance required")
```

```python
try:
    file = open("store.txt")

except FileNotFoundError:
    print("File does not exist")
```
