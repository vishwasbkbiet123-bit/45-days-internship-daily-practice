## Seaborn Line Plot :-

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
var = [1,2,3,4,5,6,7]

var_1 = [2,3,4,1,5,6,7]
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
x_1 = pd.DataFrame({"var":var, "var_1":var_1})
```

```python
sns.lineplot(x="var",y="var_1",data=x_1)
plt.show()
```

```python
# without dataframe
sns.lineplot(x=var,y=var_1)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y_1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",
             y="bill_depth_mm",
             data=y_1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",
             y="bill_depth_mm",
             hue = "sex",
             data=y_1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",
             y="bill_depth_mm",
             hue = "sex",
             style ="sex",
             data=y_1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",
             y="bill_depth_mm",
             hue = "sex",
             style ="sex",
             palette="gist_rainbow",
             data=y_1)
plt.show()
```
