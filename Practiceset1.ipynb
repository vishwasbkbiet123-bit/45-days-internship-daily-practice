```python
# 2. Write a program to calculate and print the sum, difference, product, and division of two numbers.
try:
    num1 = float(input("Enter first number :"))
    num2 = float(input("Enter second number :"))
 
    print(f"Sum of {num1} and {num2} is        : {num1 + num2}")
    print(f"Difference of {num1} and {num2} is : {num1 - num2}")
    print(f"Product of {num1} and {num2} is    : {num1 * num2}")

    if num2 == 0:
        print("Division Cannot be Zero.")
    else:
        print(f"Division of {num1} and {num2} is   : {num1 / num2}")
except ValueError:
    print("Invalid input. Please enter a valid number")
```

```python
# 3. Write a program to convert a temperature from Celsius to Fahrenheit.
try:
    Celsius = float(input("Enter temperature in celsius :-"))

    Fahrenheit = (Celsius * (9/5)) + 32
    print(f"{Celsius}°C --> {Fahrenheit}°F ")

except ValueError:
    print("Invalid input. Please enter a valid number")
```

```python
# 4. Take the radius of a circle as input and calculate its area and circumference.

try:
    radius = float(input("enter radius:-"))
    area = (22/7)*(radius**2)
    circumference = 2*(22/7)*radius
    print(f"area          :{area:.2f}")
    print(f"circumference :{circumference:.2f}")
except ValueError:
    print("Invalid input. Please enter a valid number")
```

```python
# 6. Write a program to calculate Simple Interest.
try:
    Principal = float(input("Enter Principal :"))
    Rate = float(input("Enter Rate :"))
    Time = float(input("Enter time in years :"))
    simple_interest = (Principal*Rate*Time)/100
    print(f"Simple Interest = {simple_interest}")
except ValueError:
    print("Invalid input. Please enter a valid number")
```

```python
# 7. Write a program to calculate Compound Interest.
try:
    p = float(input("Enter Principal :"))
    r = float(input("Enter Rate in % :"))/100
    t = float(input("Enter time in years :"))
    n = int(input("Compounding frequency:"))

    amount = p * (1 + r / n) ** (n * t)  

    compound_interest = amount - p
    print(f"Total Amount      = {amount:.2f}")
    print(f"compound Interest = {compound_interest:.2f}")
except ValueError:
    print("Invalid input. Please enter a valid number")
```

```python
# 8. Take a number of days as input and convert it into Years, Weeks, and Days
try:
    total_days = int(input("Number of days:-"))

    year = total_days//365
    remaining_days = total_days % 365

    week = remaining_days//7
    Days = remaining_days%7
    print(f"{total_days} days = {year} years, {week} weeks, {Days} days")
    
except ValueError:
    print("Invalid input. Please enter a valid number")
```

```python
# 9. Take the user's weight and height as input and calculate their BMI (Body Mass Index).
try:
    weight = float(input("Enter your weight in kg :"))
    height = float(input("Enter your height in cm :"))/100

    BMI = weight/(height**2)
    print(f"BMI(Body Mass Index) = {BMI:.2f}")

except ValueError:
    print("Invalid input. Please enter a valid number")
```

```python
# 10. Calculate the volume and surface area of a cylinder.
try:
    radius = float(input("Enter radius"))
    height = float(input("Enter height"))
    V = (22/7)*(radius**2)*height
    S = (2*(22/7)*radius)*(height + radius)
    print(f"Volume of a Cylinder : {V:.2f}")
    print(f"Surface of a Cylinder : {S:.2f}")
except ValueError:
    print("Invalid input. Please enter a valid number")
```

### Control Flow - If-Else & Conditionals (Easy to Medium)

```python
# 11. Check whether a number is Even or Odd
try:
    number = int(input("Enter number :"))
    if number%2 == 0:
        print(f"{number} is even.")
    else:
        print(f"{number} is odd.")
except ValueError as e:
    print("Invalid input. Please enter a valid number")
```

```python
# 12. Check whether a number is Positive, Negative, or Zero.
try:
    number = int(input("Enter number :"))
    if number == 0:
        print(f"Number is {number}")
    if number > 0:
        print(f"{number} is Positive.")
    elif number < 0:
        print(f"{number} is Negative.")
except ValueError:
    print("Invalid input. Please enter a valid number")
```

```python
# 13. Take a year as input from the user and determine if it is a Leap Year or not.

user_input = input("Enter your year of birth :")
if(len(user_input)) == 4 and user_input.isdigit():
    year = int(user_input)
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(f"{year} is a leep year")
    else:
        print(f"{year} is not a leep.")
else:
    print("Wrong input!. Please enter 4 digit (eg:- 2026)")
```

```python
# 14. Find the largest number among three given numbers.
try:
    number1 = int(input("Enter first number:-"))
    number2 = int(input("Enter second number:-"))
    number3 = int(input("Enter third number:-"))
    if (number1 > number2) and (number1 > number3):
        print(f"{number1} is greater")
    elif (number2 > number1) and (number2 > number3):
        print(f"{number2} is greater")
    elif (number3 > number1) and (number3 > number2):
        print(f"{number3} is greater")
    elif (number1 == number2) and ((number1 and number2) > number3):
        print(f"{number1} and {number2} are equal and greater")  
    elif (number2 == number3) and ((number2 and number3) > number1):
        print(f"{number2} and {number3} are equal and greater")
    elif (number1 == number3) and ((number1 and number3) > number2):
        print(f"{number1} and {number3} are equal and greater")
    else:
        print(f"{number1} = {number2} = {number3} : all are equal")
except ValueError:
    print("Invalid input!. Please enter valid number")
```
