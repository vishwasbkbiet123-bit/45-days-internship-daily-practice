## Broadcasting in Numpy :-

```python
import numpy as np
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])
print(var1 + var2)
```

```python
var1 = np.array([1,2,3])
print(var1.shape)
print()
var2 = np.array([[1],[2],[3]])
print(var2)
print()
print(var2.shape)
print()
print(var2)
print()
print(var1 + var2)
```

```python
x = np.array([[1],[2]])
print(x.shape)

y = np.array([[1,2,3],[1,2,3]])
print(y.shape)

print(x+y)
```

## Indexing Numpy Arrays :-

```python
var = np.array([9,8,7,6])

print(var[1])
print(var[-2])
```

```python
var1 = np.array([[9,8,7],[4,5,6]])

print(var1)
print()
print(var1.ndim)
print()
print(var1[0,1])
```

```python
var2 = np.array([[[1,2],[1,2]]])
print(var2)
print()
print(var2.ndim)
print()
print(var2[0,1,1])
```

```python
# Slicing

var1 = np.array([1,2,3,4])
print(var1)
print()

print(var1[1:5])

print(var1[1:])

print(var1[:5])
print(var1[::])

print(var1[1:8:2])
```

```python
var1 = np.array([[1,2,3,4,5],[9,8,7,6,5],[11,12,13,14,15]])

print(var1[1,1:])
```
