## Arithmetic Operation in Numpy Array 

```python
import numpy as np
```

```python
var = np.array([1,2,3,4])
varadd = var + 3
varsub = var - 3
varmul = var * 3
vardiv = var * 4
varmod = var % 2

print(varadd)
print(varsub)
print(varmul)
print(vardiv)
print(varmod)
```

```python
var1= np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = var1 + var2
print(varadd)
```

```python
var1= np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = np.add(var1,var2)
print(varadd)
```

function for arithmetic :--

| Function            | Description                      | Example                 |
| ------------------- | -------------------------------- | ----------------------- |
| `np.add()`          | Adds array elements              | `np.add(a, b)`          |
| `np.subtract()`     | Subtracts array elements         | `np.subtract(a, b)`     |
| `np.multiply()`     | Multiplies array elements        | `np.multiply(a, b)`     |
| `np.divide()`       | Divides array elements           | `np.divide(a, b)`       |
| `np.floor_divide()` | Returns quotient without decimal | `np.floor_divide(a, b)` |
| `np.mod()`          | Returns remainder                | `np.mod(a, b)`          |
| `np.power()`        | Raises elements to a power       | `np.power(a, b)`        |
| `np.remainder()`    | Returns division remainder       | `np.remainder(a, b)`    |
| `np.absolute()`     | Returns absolute value           | `np.absolute(a)`        |
| `np.sqrt()`         | Returns square root              | `np.sqrt(a)`            |


### 2D ARRAY :-

```python
var21 = np.array([[1,2,3,4],[1,2,3,4]])
var22 = np.array([[1,2,3,4],[1,2,3,4]])

varadd = np.add(var21,var22)
print(varadd)
```

Numpy Common Mathematic Function :-
| Function        | Description                              |
| --------------- | ---------------------------------------- |
| `np.min(x)`     | Array ka smallest value return karta hai |
| `np.max(x)`     | Array ka largest value return karta hai  |
| `np.sum(x)`     | Sabhi elements ka sum return karta hai   |
| `np.mean(x)`    | Average (mean) return karta hai          |
| `np.median(x)`  | Median value return karta hai            |
| `np.std(x)`     | Standard deviation return karta hai      |
| `np.var(x)`     | Variance return karta hai                |
| `np.argmin(x)`  | Minimum value ka index return karta hai  |
| `np.argmax(x)`  | Maximum value ka index return karta hai  |
| `np.sqrt(x)`    | Square root return karta hai             |
| `np.cumsum(x)`  | Cumulative sum return karta hai          |
| `np.cumprod(x)` | Cumulative product return karta hai      |


```python
var = np.array([2,3,4,3,5,6,3])

print("min :", np.min(var), np.argmin(var))
print("max :", np.max(var), np.argmax(var))
print("sqrt :",np.sqrt(var))
```

```python
var1 = np.array([[2,1,3],[9,5,6]])
print(np.min(var1,axis=0))
```

```python
var2 = np.array([1,2,3])
print(np.sin(var2))
print(np.cos(var2))
print(np.cumsum(var2))
```
