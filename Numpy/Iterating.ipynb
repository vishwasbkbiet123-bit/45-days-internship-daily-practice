## Iterating in Numpy :-

```python
import numpy as np
```

```python
var = np.array([9,8,7,6,5,4])
print(var)
print()
for i in var: 
    print(i)
```

```python
var1 = np.array([[1,2,3,4],[1,2,3,4]])
print(var1)
print()

for j in var1:
    print(j)
    
print()
for k in var1:
    for l in k:
        print(l)
```

```python
var3 = np.array([[[9,8,7,6],[1,2,3,4]]])
print(var3)
print()
print(var3.ndim)
print()
for i in var3:
    for j in i:
        for k in j:
            print(k)
```

```python
var3 = np.array([[[9,8,7,6],[1,2,3,4]]])
print(var3)
print()
print(var3.ndim)
print()

for i in np.nditer(var3,flags = ['buffered'], op_dtypes = ["S"]):
    print(i)
```

```python
var3 = np.array([[[9,8,7,6],[1,2,3,4]]])
print(var3)
print()
print(var3.ndim)
print()

for i in np.nditer(var3):
    print(i)
```

```python
var3 = np.array([[[9,8,7,6],[1,2,3,4]]])
print(var3)
print()
print(var3.ndim)
print()

for i,d in np.ndenumerate(var3):
    print(i,d)
```

## Copy vs Views Numpy Python Array :-

```python
var = np.array([1,2,3,4])

co =  var.copy()

var[1]  = 40
print("var :",var)
print("copy :",co)

```

```python
x = np.array([9,8,7,6,5])

xi = x.view()
x[1] = 40
print("x :",x)
print("view :",xi)
```

### Difference between Copy and View :-

| Feature            | `copy()`                                 | `view()`                        |
| ------------------ | ---------------------------------------- | ------------------------------- |
| Memory             | Creates a new memory location            | Shares the same memory location |
| Data Change Effect | Changes do **not** affect original array | Changes affect original array   |
| Speed              | Slower (copies data)                     | Faster (no data copy)           |
| Storage            | Uses extra memory                        | Uses less memory                |
| Object             | Independent array                        | Reference of original array     |
| Syntax             | `arr.copy()`                             | `arr.view()`                    |


## Join and Split function Numpy Array :-

##### join() :- Used to combine two or more arrays into a single array.
##### split() :- Used to divide a single array into multiple smaller arrays.

```python
# to add two array
var = np.array([1,2,3,4,5])
var1 = np.array([6,7,8,9])

ar = np.concatenate((var,var1))
print(ar)
```

```python
# to add 2D array 
vr = np.array([[1,2],[3,4]])
vr1 = np.array([[6,7],[8,9]])

ar_new = np.concatenate((vr,vr1),axis = 0)
print(vr)
print()
print(vr1)
print()
print(ar_new)
```

```python
vr = np.array([[1,2],[3,4]])
vr1 = np.array([[6,7],[8,9]])

ar_new = np.concatenate((vr,vr1),axis = 1)
print(vr)
print()
print(vr1)
print()
print(ar_new)
```

```python
vr_1 = np.array([[1,2,3,4]])
vr_2 = np.array([[6,7,8,9]])

a_new = np.stack((vr_1,vr_2),axis = 1)

print(a_new)
```

```python
vr_1 = np.array([[1,2,3,4]])
vr_2 = np.array([[6,7,8,9]])

a_new = np.hstack((vr_1,vr_2))

print(a_new)
```

```python
vr_1 = np.array([[1,2,3,4]])
vr_2 = np.array([[6,7,8,9]])

a_new = np.vstack((vr_1,vr_2))

print(a_new)
```

```python
vr_1 = np.array([[1,2,3,4]])
vr_2 = np.array([[6,7,8,9]])

a_new = np.dstack((vr_1,vr_2))

print(a_new)
```

```python
# split 
varr = np.array([1,2,3,4])
print(varr)

ar = np.array_split(varr,3)
print()
print(ar)
```

```python
vara = np.array([[1,2],[3,4],[5,6]])
print(vara)

ar = np.array_split(vara,3)
ar_n = np.array_split(vara,3,axis=1)
print()
print(ar)
print()
print(ar_n)
```
