## Create Numpy Arrays With Random Numbers

NumPy Random Array Functions :-

rand() → Generates random decimal numbers between 0 and 1.


randn() → Generates random numbers from a normal (Gaussian) distribution.


ranf() → Generates random floating-point numbers between 0 and 1.


randint() → Generates random integers within a specified range.

```python
# rand(0)
import numpy as np

var = np.random.rand(4)

print(var)
```

```python
# for 2D :-

var = np.random.rand(4,4)

print(var)
```

```python
# randn() :-
var2 = np.random.randn(5)
print(var2)
```

```python
# ranf() :-
var3 = np.random.ranf(4)
print(var3)
```

```python
# randint() :- 
var4 = np.random.randint(5, 20, 5)
print(var4)
```
