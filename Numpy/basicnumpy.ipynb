## NUMPY IN PYTHON ---->

What is numpy..?

NumPy is a Python library that provides support for multidimensional arrays and various mathematical functions to perform numerical computations efficiently.

Why use numpy..?

•  Faster than Python lists


•  Uses less memory


•  Supports multi-dimensional arrays


•  Provides many mathematical functions


•  Widely used in Data Science, Machine Learning, and AI


What is numpy array...?

A NumPy array is a collection of elements of the same data type stored in a contiguous memory location.




Types of numpy array :-

There are three main types of NumPy arrays:


1D Array – Single row of elements.


2D Array – Rows and columns (matrix).


3D Array – Collection of multiple 2D arrays.

Differnce between numpy array and list in python :-

| Feature                     | Python List                    | NumPy Array                        |
| --------------------------- | ------------------------------ | ---------------------------------- |
| Library Required            | No                             | Yes (`import numpy`)               |
| Speed                       | Slower                         | Faster                             |
| Memory Usage                | More memory                    | Less memory                        |
| Data Types                  | Can store different data types | Usually stores same data type      |
| Mathematical Operations     | Limited                        | Easy and efficient                 |
| Performance with Large Data | Slower                         | Much faster                        |
| Used For                    | General-purpose storage        | Numerical and scientific computing |
