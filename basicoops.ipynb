### OOPS :-->

OOP is a way of writing computer programs by organizing everything into "objects" that represent real-world things.

Core Definitions :-

→ Class: A blueprint or template for creating objects. (e.g., A car's manufacturing blueprint).

→ Object: A real, physical entity created from the class blueprint. (e.g., Your specific red Honda Civic).

→ Attributes (Properties): Variables inside a class that hold data. (e.g., color, brand, top_speed).

→ Methods: Functions inside a class that define the object's behavior. (e.g., start_engine(), brake()).

→ Constructor (__init__): A special method that automatically runs when you create a new object. It sets up the initial state.

→ self: A keyword that refers to the specific object being created or used right now.

WHY OOPS ..?

• Reusable code
• Easy maintenance
• Better organization
• Real-world modeling

### Encapsulation :-
Encapsulation in Python is the practice of bundling data (attributes) and methods (functions) into a single unit (a class) while restricting direct access to the internal data

In Python, we place underscores (_) before variables to hide data. There are three levels of this:
1. Public: Open to everyone. (No underscore: variable)
2. Protected: Accessible only to the class and its subclasses. (Single underscore: _variable) 
3. Private: Strictly secret; accessible only within the class itself. (Double underscore: __variable) 

```python
# Encapsulation :- Wrapping data and methods together inside a class is called Encapsulation

class Student:                             # create class
    def __init__(self,name):               # create constructor    
        self.name = name                   # create instance variable
```

```python
# example of encapsulation:- 
# public
class Book:
    def __init__(self, title, price):
        # Anyone can see and change these
        self.title = title        
        self.price = price        

    # Anyone can trigger this action
    def show_info(self):
        return f"Book: {self.title} costs ₹{self.price}"

# 1. Buying a book
my_book = Book("Harry Potter", 500)

# 2. Reading the data directly (Public Access)
print(my_book.title)  

# 3. Changing the price directly (No restrictions)
my_book.price = 350   
print(my_book.show_info())  
```

```python
# private
class Book:
    def __init__(self, title, price):
        self.title = title        # Public variable
        self.__price = price      # Private variable (Hidden with __)

    # 1. Getter Method: The only way to "see" the price from outside
    def get_price(self):
        return self.__price

    # 2. Setter Method: The only way to "change" the price safely
    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price
        else:
            print("Error: Price cannot be zero or negative!")

# Creating a book object
my_book = Book("Harry Potter", 500)

# 1. Checking the price safely
print(my_book.get_price())  

# 2. Trying to set a wrong price (Blocked by our rules)
my_book.set_price(-200)    

# 3. Setting a correct price (Allowed)
my_book.set_price(450)
print(my_book.get_price())  
```

```python
# protected 
# Parent Class
class SmartPhone:
    def __init__(self, brand, os):
        self.brand = brand      # Public
        self._os = os           # Protected (Shared only with family/children)

# Child Class (Inherits from SmartPhone)
class Samsung(SmartPhone):
    def __init__(self, brand, os, model):
        super().__init__(brand, os)
        self.model = model

    def show_details(self):
        # Child class can freely access the protected self._os
        return f"{self.brand} {self.model} runs on {self._os}"

# Creating a child object
my_phone = Samsung("Samsung", "Android", "Galaxy S24")

# 1. Inside access: Child class uses it successfully
print(my_phone.show_details())  

# 2. Outside access: This works, but it breaks the programmer's rule!
print(my_phone._os)             
```

```python
# protected 
# Parent Class (The Household)
class ParentHouse:
    def __init__(self, family_name):
        self.family_name = family_name  # Public: Everyone knows the family name
        self._netflix_password = "SecretFamilyPass123"  # Protected: For family only

# Child Class (Inherits everything from the parent)
class Child(ParentHouse):
    def __init__(self, family_name, child_name):
        super().__init__(family_name)
        self.child_name = child_name

    # The child can freely access the protected family password
    def watch_movie(self):
        return f"{self.child_name} logged into Netflix using password: {self._netflix_password}"

# 1. Setting up the family and the child
kid = Child(family_name="Sharma", child_name="Rahul")

# 2. Inside the family: The child uses the password successfully
print(kid.watch_movie())  

# 3. Outside the family: An outsider (stranger) tries to access it
# Python allows this, but it breaks the "social agreement" of the code!
print(kid._netflix_password)  
```

### INHERITANCE :-

Inheritance :-
Inheritance is an OOP feature that allows one class (child class) to acquire the properties and methods of another class (parent class), enabling code reusability and hierarchical relationships.

Parent Class (Base Class): A class whose properties and methods are inherited by another class.

Child Class (Derived Class): A class that inherits properties and methods from a parent class and can also add its own features.

Types of Inheritance :-

• Single Inheritance:- A derived class is created from a single base class.

• Multiple Inheritance:- A derived class is created from more than one base class.

• Multilevel Inheritance:- A derived class is created from another derived class (acting as a base class).

• Hierarchical Inheritance:- Multiple derived classes are created from a single base class.

• Hybrid Inheritance:- A combination of two or more types of inheritance within a program.

• super() :- super() ek built-in function hai jo parent class ke methods aur constructor ko child class se access karne ke liye use hota hai.

             syntax :- super().method_name()


```python
# Single Inheritance 
# Parent Class
class Animal:
    def eat(self):
        print("Eating...")

# Child Class (Animal se inherit kar rahi hai)
class Dog(Animal):
    def bark(self):
        print("Bhow! Bhow!")

# Dog ka ek object banaya
my_dog = Dog()

# 1. Parent (Animal) ka function use kiya
my_dog.eat()

# 2. Apna (Dog) ka function use kiya
my_dog.bark()
```

```python
# Multiple inheritance :-

# Parent Class 1
class Mother:
    def sing(self):
        print("Mummy achha gaati hain!")

# Parent Class 2
class Father:
    def drive(self):
        print("Papa gaadi chalate hain!")

# Child Class (Dono se inherit kar rahi hai)
class Child(Mother, Father):
    def play(self):
        print("Baccha game khelta hai!")

# --- Code Run Karna ---

# Child ka ek object banaya
kid = Child()

# 1. Mother ka function use kiya
kid.sing()   

# 2. Father ka function use kiya
kid.drive()  

# 3. Apna (Child) ka function use kiya
kid.play()
```

```python
# Multilevel inheritance :-

# Level 1: Grandparent Class
class Grandfather:
    def farming(self):
        print("Dada ji kheti karte hain!")

# Level 2: Parent Class (Grandfather se inherit kar raha hai)
class Father(Grandfather):
    def business(self):
        print("Papa business sambhalte hain!")

# Level 3: Child Class (Father se inherit kar raha hai)
class Son(Father):
    def coding(self):
        print("Beta Python me coding karta hai!")

# Sabse neeche wale (Son) ka ek object banaya
my_son = Son()

# 1. Dada ji ka function automatically mil gaya!
my_son.farming()   

# 2. Papa ka function bhi mil gaya!
my_son.business()  

# 3. Apna (Son) ka khud ka function
my_son.coding()
```

```python
# Single Parent Class
class Father:
    def surname(self):
        print("Surname: Sharma")

# Child Class 1 (Father se inherit kar rahi hai)
class Son(Father):
    def cricket(self):
        print("Beta cricket khelta hai!")

# Child Class 2 (Yeh BHI Father se inherit kar rahi hai)
class Daughter(Father):
    def dance(self):
        print("Beti dance karti hai!")

# 1. Bete (Son) ke paas Papa ka surname aur apna cricket hai
my_son = Son()
my_son.surname()   # Papa ka function
my_son.cricket()   # Apna function

print("-" * 30)

# 2. Beti (Daughter) ke paas bhi Papa ka surname aur apna dance hai
my_daughter = Daughter()
my_daughter.surname() # Papa ka function
my_daughter.dance()   # Apna function
```

```python
# Hybrid Inheritance :-

# 1. Sabse Upar Wali Parent Class
class College:
    def name(self):
        print("College: BKBIET")

# 2. Hierarchical Inheritance: College se Student aur Teacher bane
class Student(College):
    def study(self):
        print("Student: Padhai karta hai.")

class Teacher(College):
    def teach(self):
        print("Teacher: Padhate hain.")

# 3. Multiple Inheritance: Student aur Teacher dono se milkar TA bana
class TeachingAssistant(Student, Teacher):
    def assist(self):
        print("TA: Padhai bhi karta hai aur padhata bhi hai!")

# --- Code Run Karna ---

# Ek TA (Teaching Assistant) ka object banaya
ta = TeachingAssistant()

# College (Grandparent) ka function
ta.name()   

# Student (Parent 1) ka function
ta.study()  

# Teacher (Parent 2) ka function
ta.teach()  

# Apna khud ka function
ta.assist()
```

### POLYMORPHISM :- 

Polymorphism (Poly = Many, Morph = Forms) is a core concept in Object-Oriented Programming (OOP) that describes the ability of a single function, method, or operator to behave differently based on the object it is acting upon or the context in which it is used.

Two Types of Polymorphism :
1. Complile time Polymorphism

    a. method overloading 


    b. operator overloading


2. Run time polymorphism :
    
    
     method overriding

```python
# Compile time polymorphism:-

class Area:
    # Ek hi method jo alag-alag arguments par alag kaam karega
    def calculate(self, length=None, breadth=None):
        # Agar sirf length di hai, toh square ka area
        if length is not None and breadth is None:
            return length * length
        # Agar length aur breadth dono di hain, toh rectangle ka area
        elif length is not None and breadth is not None:
            return length * breadth
        else:
            return "Khaali parameter nahi chalega!"

# Object create karna
obj = Area()

# Ek argument ke sath call (Square)
print(obj.calculate(5))      

# Do arguments ke sath call (Rectangle)
print(obj.calculate(5, 10))  
```

```python
# operator overloading 
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    # '+' operator ko overload karne ke liye dunder method
    def __add__(self, other):
        # Naya real aur imaginary part calculate karna
        total_real = self.real + other.real
        total_imag = self.imag + other.imag
        # Ek naya object return karna
        return ComplexNumber(total_real, total_imag)

    # Output ko readable format me print karne ke liye
    def __str__(self):
        return f"{self.real} + {self.imag}i"

# Do alag-alag objects banana
num1 = ComplexNumber(3, 4)  
num2 = ComplexNumber(1, 2)  

result = num1 + num2

print("Result:", result)  
```

```python
# run time polymorphism 

# Parent Class
class Animal:
    def sound(self):
        return "Kuch aawaz"

# Child Class 1 (Override kiya)
class Dog(Animal):
    def sound(self):
        return "Bhow Bhow"

# Child Class 2 (Override kiya)
class Cat(Animal):
    def sound(self):
        return "Meow"

# Run-time par check karna
def make_sound(animal_obj):
    print(animal_obj.sound()) # Run-time par decide hoga

# Alag-alag objects bhejna
make_sound(Dog())  
make_sound(Cat())  
```

### Abstraction :- 

Abstraction is the process of hiding implementation details and showing only the essential features of an object.

Example:
When you drive a car, you use the steering wheel, brake, and accelerator, but you don't need to know how the engine works internally. This is abstraction.



```python
# Abstraction :-

class ATM:
    def check_balance(self):
        print("Balance: ₹10,000")

    def withdraw(self, amount):
        print(f"₹{amount} Withdrawn Successfully")

atm = ATM()

atm.check_balance()
atm.withdraw(2000)

# User only uses check_balance() and withdraw() methods. The internal process (database, verification, etc.) is hidden from the user.
```
