→ Data Visualization is the graphical representation of information and data.

→ visual elements like charts, graphs, and maps, data visualization tools provide an accessible     way to see and understand trends, outliers, and patterns in data.

## Matplotlib Bar Plot :-

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# DATA
x = ["Python","c","c++","java"]
y =[85,70,60,82]

# Axis labels and title
plt.xlabel("language",fontsize=15)
plt.ylabel("no.",fontsize=15)
plt.title("BKBIET",fontsize=15)

# Different colors for each bar
c = ["y","b","r","g"]

# Create bar graph
plt.bar(x,y,
        width=0.4,
        color=c,
        align="center"
        )

# Display Graph
plt.show()
```

```python
# data
x = ["Python","c","c++","java"]
y =[85,70,60,82]

# axis label and title
plt.xlabel("language",fontsize=15)
plt.ylabel("no.",fontsize=15)
plt.title("BKBIET",fontsize=15)

# color for each bar
c = ["y","b","r","g"]

# create bar graph with boder and linestyle
plt.bar(x,y,
        width=0.4,
        color=c,
        align="center",
        edgecolor="orange",
        linewidth=5,
        linestyle=":"
        )

# display graph
plt.show()
```

```python

x = ["Python","c","c++","java"]
y =[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("no.",fontsize=15)
plt.title("BKBIET",fontsize=15)
c = ["y","b","r","g"]
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="orange",linewidth=5,linestyle=":",alpha=0.5)
plt.show()
```

```python
# data
x = ["Python","c","c++","java"]
y =[85,70,60,82]

# axis label with title
plt.xlabel("language",fontsize=15)
plt.ylabel("no.",fontsize=15)
plt.title("BKBIET",fontsize=15)

# create bar graph 
plt.bar(x,y,
        width=0.4,
        color="r",
        align="center",
        edgecolor="orange",
        label="Popularity"
        )

# show legand
plt.legend()

# show display 
plt.show()
```

```python
# data
x = ["Python","c","c++","java"]
y =[85,70,60,82]
z =[20,30,40,50]

# Bar width and positions
width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

# Labels and title
plt.xlabel("language",fontsize=15)
plt.ylabel("no.",fontsize=15)
plt.title("BKBIET",fontsize=15)

# First bar graph
plt.bar(p,y,width,color="r",align="center",edgecolor="orange",label="Popularity")

# second bar graph 
plt.bar(p1,z,width,color="y",align="center",edgecolor="orange",label="Popularity1")

# Set X-axis labels and rotate them
plt.xticks(p+width/2,x, rotation=10)
# show legand 
plt.legend()

# show graph 
plt.show()
```

```python
# data
x = ["Python","c","c++","java"]
y =[85,70,60,82]
z =[20,30,40,50]

# Labels and title
plt.xlabel("language",fontsize=15)
plt.ylabel("no.",fontsize=15)
plt.title("BKBIET",fontsize=15)

# First bar graph
plt.barh(p,y,color="r",align="center",edgecolor="orange",label="Popularity")

# second bar graph 
plt.barh(p1,z,color="y",align="center",edgecolor="orange",label="Popularity1")

# show legand 
plt.legend()

# show graph 
plt.show()
```

##  Matplotlib Scatter Plot :-

```python
# data
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

# create scatter plot
plt.scatter(day,no,color="r")

# axis labeling and title
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)

# display graph
plt.show()
```

```python
# data
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

colors = ["r","y","g","b","r","g","r"]
sizes = [400,200,400,300,200,100,600]
# create scatter plot
plt.scatter(day,no,c=colors,s=sizes)

# axis labeling and title
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)

# display graph
plt.show()
```

```python
# data
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

colors = ["r","y","g","b","r","g","r"]
sizes = [400,200,400,300,200,100,600]
# create scatter plot
plt.scatter(day,no,c=colors,s=sizes)

# axis labeling and title
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)

# display graph
plt.show()
```

```python
# data
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

colors = ["r","y","g","b","r","g","r"]
sizes = [400,200,400,300,200,100,600]
# create scatter plot
plt.scatter(day,no,c=colors,s=sizes,marker="*",edgecolor="g",linewidth=2)

# axis labeling and title
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)

# display graph
plt.show()
```

```python
# data
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

colors = [10,49,30,29,56,20,30]
sizes = [400,200,400,300,200,100,600]
# create scatter plot
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
t = plt.colorbar()
t.set_label("ColorBar",fontsize=15)
# axis labeling and title
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)

# display graph
plt.show()
```
