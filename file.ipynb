#### FILE HANDLING -->

```python
"""
What is File Handling :- 
File Handling is the process of creating, reading, writing, updating, and deleting files using Python. 
"""
```

```python
# Opening a File :- Before working with a file, we must open it.
"""
Syntax -->
open("filename", "mode")
"""
file = open("list.ipynb", "r")
file.close()
```

```python
file = open("data.txt", "r") # use to read data from a file
```

```python
file = open("data.txt", "w") # use to write data into a file
```

```python
file = open("data.txt", "a") # use to add new data at the end of a file
```

```python
file.close() # closing releases the resources used by the file
```

```python
# Reading data 

file = open("data.txt", "r")

data = file.read()

print(data)

file.close()
```

```python
file = open("data.txt", "r")

file.read(5)
```

```python
file.readline()
```

```python
file.readlines()
```

```python
# writing data

file = open("data.txt", "w")
file.write("Hello Python")
file.close()
```

```python
# Appending Data

file = open("data.txt", "a")
file.write("\nNew Line")
file.close()
```

```python
# with statement

with open("data.txt", "r") as file:
    print(file.read())
```

```python
# File Pointer:- A file pointer is the current position of the cursor inside a file.
"""
Hello Python
^
Cursor starts at beginning.

After reading:
Hello Python
            ^
Cursor moves forward.    
"""
```

```python
# tell :- Returns current cursor position
file = open("data.txt","r")
file.tell()
```

```python
# Working with Text Files (.txt)

with open("notes.txt","r") as file:
    print(file.read())
```
