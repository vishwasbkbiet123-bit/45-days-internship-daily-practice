```python
""" 
1. Explain the difference between the following data types with examples:
• Integer (int)
• Float (float)
• String (str)
• Boolean (bool)
"""
#Integer (int): A data type used to store whole numbers without decimal points.
# Example: 10, -5, 100
num = 100
print("int :-",num)

# Float (float): A data type used to store numbers with decimal points.
# Example: 3.14, 99.9, -2.5
salary = 25000.50
print("Float :-",salary)

# String (str): A data type used to store text enclosed in quotes.
# Example: "Hello", 'Python'
name = "Vishwas"
print("string :-",name)

# Boolean (bool): A data type used to store logical values, either True or False.
# Example: True, False
is_passed = True
print("bool :-",is_passed)
```

```python
"""
3. Write a Python program that:
Takes a user's name as input.
Prints the name in uppercase.
Prints the total number of characters in the name.
"""
name = input("Enter your name: ")

print("Uppercase Name:", name.upper())
print("Total Characters:", len(name))
```

```python
""" 
4. Explain any five commonly used string methods in Python with
examples.
"""
# 1. upper()
# Converts all characters of a string to uppercase.
name = "vishwas"
print(name.upper())

# 2. lower()
# Converts all characters of a string to lowercase.
name = "VISHWAS"
print(name.lower())

# 3. title()
# Converts the first letter of each word to uppercase
text = "hello world"
print(text.title())

# 4. replace()
# Replaces a specific word or character with another.
text = "I like Java"
print(text.replace("Java", "Python"))

# 5. len()
# Returns the total number of characters in a string.
name = "Vishwas"
print(len(name))
```

```python
""" 
5. Create a list containing the names of five fruits.
• Print the complete list.
• Print the first and last element.
• Print the total number of items in the list.
"""

fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

print("Complete List:", fruits)
print("First Element:", fruits[0])
print("Last Element:", fruits[-1])
print("Total Number of Items:", len(fruits))
```
